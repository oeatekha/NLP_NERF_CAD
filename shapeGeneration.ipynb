**Generating Data**

For this project, we have to generate large sums of data for the cylinder, sphere, and cubic shapes. The following will introduce a funciton to create a random dataset.

```python
import numpy as np
import pandas as pd
import math
import random


token_0 = "[CLS]"
numberOfExamples = 10000
shapes = ["Sphere", "Cylinder", "Cubic"]
generatedShape = []
#sentence = "[CLS] Shape { 23.2,13,14.2 }"

def createSentece(numberOfExamples):
    shapeList = []

    for i in range(numberOfExamples):
        shape = random.choice(shapes)
        
        sentence = ""

        if shape == "Sphere":
            radius = np.random.uniform(0, 100).__round__(2)
            sentence = "[CLS] " + shape + " { " + str(radius) + " }" + "0 0"
        elif shape == "Cylinder":
            radius = np.random.uniform(0, 100).__round__(2)
            height = np.random.uniform(0, 100).__round__(2)
            sentence =  "[CLS] " + shape + " { " + str(radius) + ", " + str(height) + " }" + "0"
        elif shape == "Cubic":
            length = np.random.uniform(0, 100).__round__(2)
            height = np.random.uniform(0, 100).__round__(2)
            width = np.random.uniform(0, 100).__round__(2)
            sentence =  "[CLS] " + shape + " { " + str(length) + ", " + str(height) + ", " + str(width) + " }"
        else:
            print("Error")

        shapeList.append(sentence)
        #print(sentence)
    return shapeList

generatedShape = createSentece(numberOfExamples)
print(generatedShape)
```
